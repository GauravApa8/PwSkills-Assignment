6.  What are the different types of sampling methods?  
Ans.  
Sampling methods can be broadly categorized into two main types: probability sampling and non-probability sampling. Probability sampling uses random selection, giving each member of the population a known chance of being chosen. Non-probability sampling, on the other hand, relies on subjective or non-random selection methods.  
 1.   Probability Sampling Methods:
 *   Simple Random Sampling: Each individual has an equal and random chance of being selected.
 *   Systematic Sampling: Elements are selected at regular intervals from the population.
 *   Stratified Sampling: The population is divided into subgroups (strata), and samples are drawn from each.
 *   Cluster Sampling: The population is divided into clusters, and then entire clusters are randomly selected.
 2.   Non-Probability Sampling Methods:  
 *   Convenience Sampling:
Participants are selected based on their ease of access and availability.
 *   Purposive Sampling:
Participants are chosen based on specific criteria or characteristics relevant to the research.
 *   Snowball Sampling:
Participants are asked to refer other potential participants, often useful for finding hidden or hard-to-reach populations.
 *   Quota Sampling:
The population is divided into subgroups, and quotas are set for the number of participants needed from each.

10. Explain nominal, ordinal, interval, and ratio levels of measurement.

 1.   Nominal:
 *   Description: Data is categorized into distinct, unordered groups or labels.
 *   Example: Colors of cars (red, blue, green), types of fruit (apple, banana, orange).

 2.   Ordinal:

 *   Example: Temperature in Celsius or Fahrenheit, calendar years.
 *   Key feature: Equal intervals, but a zero point does not indicate the absence of the quantity.
 4.   Ratio:
 *   Description: Data is categorized, ranked, has equal intervals, and has a true zero point.
 *   Example: Height, weight, age, income, number of items sold.
 *   Key feature: Possesses all properties of interval data, plus a true zero point, allowing for meaningful ratios.

12. Define mean, median, and mode.  
Ans.  
Mean, median, and mode are measures of central tendency used to describe the "average" or typical value within a dataset. The mean is the arithmetic average, the median is the middle value when the data is ordered, and the mode is the most frequent value.  
*   Mean: The mean is calculated by summing all the values in a dataset and then dividing by the total number of values. For example, the mean of the set {2, 4, 6, 8} is (2+4+6+8)/4 = 5.

*   Median: The median is the middle value of a dataset when it is arranged in ascending order. If there is an even number of values, the median is the average of the two middle values. For example, the median of the set {2, 4, 6, 8, 10} is 6. For the set {2, 4, 6, 8}, the median is (4+6)/2 = 5.
*   Mode: The mode is the value that appears most frequently in a dataset. For example, the mode of the set {2, 2, 4, 6, 8, 8, 8} is 8. A dataset can have one mode, multiple modes, or no mode at all.

14. What is variance, and how is it calculated?  
Ans.  
Variance is a statistical measure that quantifies the amount of dispersion or variability within a set of data points. It essentially tells you how spread out the numbers are from their average value (mean). A higher variance indicates that the data points are more scattered, while a lower variance suggests they are clustered closer to the mean.   
Calculation:  
 1.   Calculate the mean: Sum all the data points and divide by the total number of points.
 2.   Find the differences from the mean: Subtract the mean from each individual data point.
 3.   Square the differences: Square each of the differences calculated in the previous step.
 4.   Calculate the average of the squared differences: Sum all the squared differences and divide by the total number of data points (or, for a sample, divide by the number of data points minus one). This final result is the variance.

16.  Define and explain the term range in statistics.

Explanation:

 2.   How to calculate it:

17.  What is the difference between variance and standard deviation?  
Ans.  
Variance and standard deviation are both measures of data dispersion, but they differ in how they quantify it and in their units. Variance is the average of the squared differences from the mean, while standard deviation is the square root of the variance. This means standard deviation is expressed in the same units as the original data, making it easier to interpret, whereas variance is in squared units.
Key Differences:  
 *   Calculation:
Variance is calculated by averaging the squared differences between each data point and the mean. Standard deviation is then calculated by taking the square root of the variance.
 *   Units:
Variance is in squared units of the original data (e.g., if the data is in meters, variance is in square meters). Standard deviation is in the same units as the original data (e.g., meters).
 *   Interpretation:
Standard deviation is generally easier to interpret because it is in the original units of the data, making it more intuitive to understand the spread of the data.
 *   In essence: Variance provides a measure of the average squared distance from the mean, while standard deviation provides a measure of the average distance from the mean, in the original units of the data.

 1.   Positively Skewed (Right Skewed):
 *   The majority of the data points are clustered on the left side of the distribution, with a long tail extending towards the right.

 *   Think of it like a graph where most of the data is on the lower end, and there are a few very high outliers.
 2.   Negatively Skewed (Left Skewed):
 *   The majority of the data points are clustered on the right side of the distribution, with a long tail extending towards the left.

 *   Imagine a graph where most data is on the higher end, with a few very low outliers.

23.  What is the difference between covariance and correlation?   
Ans.  
Covariance and correlation both measure the relationship between two variables, but they differ in how they represent the strength and scale of that relationship. Covariance indicates the direction of the relationship (positive or negative), while correlation standardizes this measure to provide a value between -1 and +1, indicating both direction and strength.   
Here's a more detailed breakdown:  
 1.   Covariance:
 *   Definition: Measures how much two random variables change together.
 *   Range: Can range from negative infinity to positive infinity (-∞ to +∞).
 *   Interpretation: Positive covariance indicates that the variables tend to move in the same direction (both increase or both decrease), while negative covariance indicates they move in opposite directions.
 *   Limitations: The magnitude of covariance is not easily interpretable because it is affected by the scales of the variables. For example, covariance can be heavily influenced by changes in the units of measurement (e.g., changing from kilograms to grams).
 2.   Correlation:
 *   Definition: A standardized measure of the linear relationship between two variables.
 *   Range: Always falls between -1 and +1 (-1 to +1).
 *   Interpretation: Similar to covariance, positive correlation indicates a positive relationship, negative correlation indicates a negative relationship, and a correlation of 0 indicates no linear relationship. However, correlation also indicates the strength of the relationship:   
  a) A correlation of +1 or -1 indicates a perfect positive or negative linear relationship, respectively.   
  b) A correlation closer to +1 or -1 indicates a stronger relationship, while a correlation closer to 0 indicates a weaker relationship.
 *   Advantages: Correlation is easier to interpret than covariance because it is unit-free and always falls within a defined range.

```python
# 4.  Implement sampling techniques like random sampling and stratified sampling.

import random
import pandas as pd

def random_sampling(population, sample_size):
    if sample_size > len(population):
      raise ValueError("Sample size cannot be greater than population size")
    return random.sample(population, sample_size)


def stratified_sampling(population_df, group_column, sample_size_per_group):
    sampled_data = []
    for group, data in population_df.groupby(group_column):
        if len(data) < sample_size_per_group:
            sampled_data.extend(data.to_dict('records'))
        else:
            sampled_data.extend(data.sample(n=sample_size_per_group, random_state=42).to_dict('records'))


    return pd.DataFrame(sampled_data)

if __name__ == '__main__':
    population = list(range(1, 101))
    sample_size = 10
    random_sample = random_sampling(population, sample_size)
    print(f"Random Sample: {random_sample}")

    data = {'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C'],
            'Value': [10, 20, 30, 15, 25, 35, 12, 22, 32, 42, 52]}
    df = pd.DataFrame(data)
    group_column = 'Category'
    sample_size_per_group = 2

    stratified_sample_df = stratified_sampling(df, group_column, sample_size_per_group)
    print("\nStratified Sample:")
    print(stratified_sample_df)
```

```python
# 8. Generate a dataset and demonstrate positive and negative skewness.

import numpy as np
from scipy.stats import skew
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# 1. Generate a positively skewed dataset (e.g., exponential distribution)
# An exponential distribution naturally exhibits positive skewness.
positive_skew_data = np.random.exponential(scale=2, size=1000)

# 2. Generate a negatively skewed dataset
# This can be achieved by subtracting a positively skewed distribution from a constant,
# or by using a beta distribution with appropriate parameters.
# Here, we'll create a "reversed" exponential-like distribution.
negative_skew_data = 10 - np.random.exponential(scale=2, size=1000)
# Ensure values are non-negative if desired, although skewness itself doesn't depend on this.
negative_skew_data[negative_skew_data < 0] = 0.1 # Small positive value to avoid zero/negative if needed for some analyses

# 3. Calculate skewness for both datasets
skewness_positive = skew(positive_skew_data)
skewness_negative = skew(negative_skew_data)

print(f"Skewness of positively skewed data: {skewness_positive:.4f}")
print(f"Skewness of negatively skewed data: {skewness_negative:.4f}")

# 4. Visualize the distributions
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(positive_skew_data, kde=True, color='skyblue')
plt.title(f'Positively Skewed Data (Skew: {skewness_positive:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(negative_skew_data, kde=True, color='lightcoral')
plt.title(f'Negatively Skewed Data (Skew: {skewness_negative:.2f})')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 9.  Write a Python script to calculate covariance between two datasets.

import numpy as np

def calculate_covariance(data1, data2):
    if len(data1) != len(data2):
        print("Error: Datasets must have the same length to calculate covariance.")
        return None

    # Convert lists to numpy arrays for efficient calculations if not already
    data1 = np.array(data1)
    data2 = np.array(data2)

    # Calculate the covariance matrix
    covariance_matrix = np.cov(data1, data2)

    # The covariance between data1 and data2 is at index [0, 1] (or [1, 0])
    covariance = covariance_matrix[0, 1]

    return covariance

dataset_a = [10, 15, 20, 25, 30]
dataset_b = [5, 8, 12, 16, 20]

cov_ab = calculate_covariance(dataset_a, dataset_b)

if cov_ab is not None:
    print(f"The covariance between dataset_a and dataset_b is: {cov_ab}")

dataset_c = [1, 2, 3]
dataset_d = [5, 6, 7, 8] # Different length

cov_cd = calculate_covariance(dataset_c, dataset_d)
if cov_cd is None:
    print("Covariance calculation failed due to unequal dataset lengths.")
```

```python
# 10. Write a Python script to calculate the correlation coefficient between two datasets.

import numpy as np

def calculate_correlation(dataset1, dataset2):
    try:
        # Convert to numpy arrays for efficient calculation
        arr1 = np.array(dataset1)
        arr2 = np.array(dataset2)

        # Check if datasets have the same length
        if len(arr1) != len(arr2):
            print("Error: Datasets must have the same length.")
            return None

        # Calculate the correlation matrix
        correlation_matrix = np.corrcoef(arr1, arr2)

        # The Pearson correlation coefficient is at index [0, 1] or [1, 0]
        correlation_coefficient = correlation_matrix[0, 1]
        return correlation_coefficient
    except ValueError as e:
        print(f"Error: Could not calculate correlation. Ensure datasets contain numeric values. {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

if __name__ == "__main__":
    # Example usage:
    data_a = [1, 2, 3, 4, 5]
    data_b = [2, 4, 5, 4, 5]

    corr = calculate_correlation(data_a, data_b)

    if corr is not None:
        print(f"The Pearson correlation coefficient between data_a and data_b is: {corr:.4f}")

    data_c = [10, 20, 30]
    data_d = [5, 10, 15, 20]

    corr_diff_len = calculate_correlation(data_c, data_d)
    if corr_diff_len is None:
        print("Correlation not calculated due to different dataset lengths.")

    data_e = [1, 2, 'a']
    data_f = [4, 5, 6]

    corr_non_numeric = calculate_correlation(data_e, data_f)
    if corr_non_numeric is None:
        print("Correlation not calculated due to non-numeric data.")
```

```python
# 12.  Implement and compare simple random sampling and systematic sampling.

import random
import numpy as np

def simple_random_sampling(population, sample_size):

    if sample_size > len(population):
        raise ValueError("Sample size cannot be greater than population size.")
    return random.sample(population, sample_size)

def systematic_sampling(population, sample_size):

    if sample_size > len(population):
        raise ValueError("Sample size cannot be greater than population size.")

    k = len(population) // sample_size  # Calculate the sampling interval
    start_index = random.randint(0, k - 1)  # Randomly choose a starting point

    sample = []
    for i in range(sample_size):
        index = start_index + i * k
        if index < len(population):  # Ensure index is within bounds
            sample.append(population[index])
        else:
            pass
    return sample


if __name__ == "__main__":
    population_size = 100
    sample_size = 10
    population = list(range(1, population_size + 1))

    print(f"Population: {population}")
    print(f"Population Size: {population_size}")
    print(f"Desired Sample Size: {sample_size}\n")

    # Simple Random Sampling
    srs_sample = simple_random_sampling(population, sample_size)
    print(f"Simple Random Sample: {srs_sample}")
    print(f"Size of Simple Random Sample: {len(srs_sample)}\n")

    # Systematic Sampling
    sys_sample = systematic_sampling(population, sample_size)
    print(f"Systematic Sample: {sys_sample}")
    print(f"Size of Systematic Sample: {len(sys_sample)}\n")
```

```python
# 13. Calculate the mean, median, and mode of grouped data.

import math
from collections import Counter

def calculate_grouped_mean(data):
    sum_fx = 0
    sum_f = 0
    for lower, upper, freq in data:
        midpoint = (lower + upper) / 2
        sum_fx += midpoint * freq
        sum_f += freq
    return sum_fx / sum_f

def calculate_grouped_median(data):
    total_frequency = sum(freq for _, _, freq in data)
    median_position = total_frequency / 2

    cumulative_frequency = 0
    median_class_lower = 0
    median_class_frequency = 0
    cumulative_frequency_before_median_class = 0
    class_width = 0

    for i, (lower, upper, freq) in enumerate(data):
        cumulative_frequency += freq
        if cumulative_frequency >= median_position:
            median_class_lower = lower
            median_class_frequency = freq
            class_width = upper - lower
            if i > 0:
                cumulative_frequency_before_median_class = sum(data[j][2] for j in range(i))
            break

    if median_class_frequency == 0:
        return float('nan') # Handle cases where median class has zero frequency

    median = median_class_lower + (class_width * (median_position - cumulative_frequency_before_median_class)) / median_class_frequency
    return median

def calculate_grouped_mode(data):
    max_frequency = 0
    modal_class = None

    for lower, upper, freq in data:
        if freq > max_frequency:
            max_frequency = freq
            modal_class = (lower, upper, freq)

    if modal_class is None:
        return float('nan') # No data provided

    mode_candidates = [(lower, upper, freq) for lower, upper, freq in data if freq == max_frequency]
    if len(mode_candidates) > 1:
        return (mode_candidates[0][0] + mode_candidates[0][1]) / 2

    lower_modal = modal_class[0]
    freq_modal = modal_class[2]
    class_width = modal_class[1] - modal_class[0]

    # Find frequencies of the class before and after the modal class
    freq_before = 0
    freq_after = 0
    for i, (lower, upper, freq) in enumerate(data):
        if (lower, upper, freq) == modal_class:
            if i > 0:
                freq_before = data[i-1][2]
            if i < len(data) - 1:
                freq_after = data[i+1][2]
            break

    delta1 = freq_modal - freq_before
    delta2 = freq_modal - freq_after

    if (delta1 + delta2) == 0:
        return (lower_modal + modal_class[1]) / 2 # Handle cases where delta1 + delta2 is zero (e.g., uniform distribution around mode)

    mode = lower_modal + (class_width * delta1) / (delta1 + delta2)
    return mode
grouped_data = [
    (10, 20, 5),
    (20, 30, 8),
    (30, 40, 12),
    (40, 50, 7),
    (50, 60, 3)
]

mean = calculate_grouped_mean(grouped_data)
median = calculate_grouped_median(grouped_data)
mode = calculate_grouped_mode(grouped_data)

print(f"Mean of grouped data: {mean:.2f}")
print(f"Median of grouped data: {median:.2f}")
print(f"Mode of grouped data: {mode:.2f}")
```

```python
# 14. Simulate data using Python and calculate its central tendency and dispersion.

import numpy as np
import statistics
from scipy import stats

def simulate_and_analyze_data(num_samples=100, mean=50, std_dev=10):
    # Simulate data from a normal distribution
    simulated_data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

    # Calculate Central Tendency
    data_mean = np.mean(simulated_data)
    data_median = np.median(simulated_data)
    try:
        data_mode = statistics.mode(simulated_data) # statistics.mode requires discrete data for reliable results
    except statistics.StatisticsError:
        data_mode = "No unique mode found or data is continuous." # Handle cases with no unique mode

    # Calculate Dispersion
    data_variance = np.var(simulated_data)
    data_std_dev = np.std(simulated_data)
    data_range = np.max(simulated_data) - np.min(simulated_data)

    q1 = np.percentile(simulated_data, 25)
    q3 = np.percentile(simulated_data, 75)
    data_iqr = q3 - q1

    results = {
        "simulated_data": simulated_data,
        "central_tendency": {
            "mean": data_mean,
            "median": data_median,
            "mode": data_mode
        },
        "dispersion": {
            "variance": data_variance,
            "standard_deviation": data_std_dev,
            "range": data_range,
            "interquartile_range": data_iqr
        }
    }
    return results

if __name__ == "__main__":
    analysis_results = simulate_and_analyze_data(num_samples=200, mean=75, std_dev=15)

    print("Simulated Data (first 10 samples):")
    print(analysis_results["simulated_data"][:10])
    print("\nCentral Tendency Measures:")
    print(f"Mean: {analysis_results['central_tendency']['mean']:.2f}")
    print(f"Median: {analysis_results['central_tendency']['median']:.2f}")
    print(f"Mode: {analysis_results['central_tendency']['mode']}")
    print("\nDispersion Measures:")
    print(f"Variance: {analysis_results['dispersion']['variance']:.2f}")
    print(f"Standard Deviation: {analysis_results['dispersion']['standard_deviation']:.2f}")
    print(f"Range: {analysis_results['dispersion']['range']:.2f}")
    print(f"Interquartile Range (IQR): {analysis_results['dispersion']['interquartile_range']:.2f}")
```

```python
# 18.  Implement Z-score normalization and explain its significance.

import numpy as np

def z_score_normalize(data):
    data = np.array(data)  # Ensure data is a NumPy array
    mean = np.mean(data)
    std_dev = np.std(data)

    # Handle the case of zero standard deviation to avoid division by zero
    if std_dev == 0:
        return np.zeros_like(data)
    else:
        normalized_data = (data - mean) / std_dev
        return normalized_data

data = [10, 20, 30, 40, 50]
normalized_data = z_score_normalize(data)

print(f"Original Data: {data}")
print(f"Normalized Data (Z-score): {normalized_data}")
print(f"Mean of Normalized Data: {np.mean(normalized_data):.2f}")
print(f"Standard Deviation of Normalized Data: {np.std(normalized_data):.2f}")
```

```python
# 22. Generate a dataset and implement both variance and standard deviation computations.

import math
import random

def generate_dataset(size=10, start=1, end=100):

    return [random.randint(start, end) for _ in range(size)]

def calculate_mean(data):
    if not data:
        return 0
    return sum(data) / len(data)

def calculate_variance(data):
    if not data:
        return 0
    mean = calculate_mean(data)
    squared_differences = [(x - mean) ** 2 for x in data]
    return sum(squared_differences) / len(data)

def calculate_standard_deviation(data):
    if not data:
        return 0
    variance = calculate_variance(data)
    return math.sqrt(variance)

# Generate a sample dataset
dataset = generate_dataset(size=15, start=10, end=200)
print(f"Generated Dataset: {dataset}")

# Compute and print the mean
mean_value = calculate_mean(dataset)
print(f"Mean: {mean_value:.2f}")

# Compute and print the variance
variance_value = calculate_variance(dataset)
print(f"Variance: {variance_value:.2f}")

# Compute and print the standard deviation
std_dev_value = calculate_standard_deviation(dataset)
print(f"Standard Deviation: {std_dev_value:.2f}")
```

```python
# 24.  Implement the Pearson and Spearman correlation coefficients for a dataset.

from scipy.stats import pearsonr
from scipy.stats import spearmanr
import numpy as np

def calculate_pearson_correlation(data1, data2):
    correlation, _ = pearsonr(data1, data2)
    return correlation
data_a = [1, 2, 3, 4, 5]
data_b = [5, 4, 3, 2, 1]
data_c = [1, 2, 3, 4, 5]
data_d = [2, 4, 6, 8, 10]

pearson_corr_ab = calculate_pearson_correlation(data_a, data_b)
pearson_corr_cd = calculate_pearson_correlation(data_c, data_d)
print("---Pearson Correlation Coefficient---")
print(f"Pearson correlation between data_a and data_b: {pearson_corr_ab:.2f}")
print(f"Pearson correlation between data_c and data_d: {pearson_corr_cd:.2f}")


def calculate_spearman_correlation(data1, data2):
    correlation, _ = spearmanr(data1, data2)
    return correlation

data_e = [1, 2, 3, 4, 5]
data_f = [5, 4, 3, 2, 1]
data_g = [1, 2, 3, 4, 5]
data_h = [2, 4, 6, 8, 10]

spearman_corr_ef = calculate_spearman_correlation(data_e, data_f)
spearman_corr_gh = calculate_spearman_correlation(data_g, data_h)
print("\n---Spearman Correlation Coefficient---")
print(f"Spearman correlation between data_e and data_f: {spearman_corr_ef:.2f}")
print(f"Spearman correlation between data_g and data_h: {spearman_corr_gh:.2f}")
```
