8. What is a hash table, and how does it relate to dictionaries in Python?     
Ans. A hash table is a data structure that stores key-value pairs, where each key is unique and mapped to a specific location (index) in an array using a hash function. This allows for efficient retrieval, insertion, and deletion of data. The hash function computes the index based on the key, aiming for even distribution to minimize collisions (multiple keys mapping to the same index).
In Python, dictionaries are implemented using hash tables. When a key-value pair is added to a dictionary, Python's built-in hash function calculates the hash value of the key, which determines the index where the value will be stored. This mechanism enables dictionaries to provide average-case O(1) time complexity for common operations like accessing, inserting, and deleting elements, making them highly efficient for various applications.

10. Explain why strings are immutable in Python.     
Ans. Strings are immutable in Python primarily for efficiency and safety reasons. Immutability allows for optimizations such as string interning, where identical string literals are stored only once in memory. If strings were mutable, modifying one instance would unintentionally affect all other references to the same string, leading to unpredictable behavior. Additionally, immutable strings can be used as dictionary keys, as their hash values remain constant throughout their lifetime. This property is crucial for maintaining the integrity of dictionaries, which rely on stable hash values for efficient lookups.

16. What is a nested dictionary, and give an example of its use case?  
Ans. A nested dictionary in Python is a dictionary where the values are themselves dictionaries. This allows for the creation of a hierarchical structure where data can be organized into multiple levels. A common use case for nested dictionaries is representing complex, multi-layered data, such as employee information with nested address and contact details.  
Here is a example of nested dictionary:

```python
student = {
    "name": "Gaurav Suman",
    "Course Enrolled": "Data ANalytics Course",
    "address": {
        "street": "123 Hardiya",
        "city": "Sitamarhi",
        "state": "Bihar",
        "zip": "843000"
    },
    "contact": {
        "phone": "7979038581",
        "email": "gaurav.suman@example.com"
    }
}
#In this example, the Student dictionary contains nested dictionaries for address and contact. This allows for easy organization and access of detailed information related to the students address and contact details.
print(student)
```

17. Describe the time complexity of accessing elements in a dictionary.  
Ans. Accessing an element in a dictionary using its key typically has a time complexity of "O(1)" on average, meaning it takes constant time, regardless of the dictionary's size. However, in the worst-case scenario, it can degrade to "O(n)", where n is the number of key-value pairs, if there are many hash collisions.
